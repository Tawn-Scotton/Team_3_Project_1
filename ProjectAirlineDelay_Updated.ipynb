```python
#Total # of Delayed/Cancelled/Diverted Flights for all years for all carriers
# arr_flights            float64
# arr_del15              float64
# arr_cancelled          float64
# arr_diverted           float64
# arr_del15 / arr_flights show as %
# arr_cancelled / arr_flights show as %
# arr_diverted / arr_flights show as %

#Refined entire dataset to "year","arr_flights","arr_del15","arr_cancelled","arr_diverted"
df2_AirlineDelay_Status = df2_AirlineDelay[["year","arr_flights","arr_del15","arr_cancelled","arr_diverted"]]

#Aggregated Data by Year 
df2_AirlineDelay_Status_Rollup = df2_AirlineDelay_Status.groupby(["year"]).sum()
#Calculate % amounts
df2_AirlineDelay_ArrDel15 = (df2_AirlineDelay_Status_Rollup['arr_del15'] / df2_AirlineDelay_Status_Rollup['arr_flights'])*100
df2_AirlineDelay_ArrCancel = (df2_AirlineDelay_Status_Rollup['arr_cancelled'] / df2_AirlineDelay_Status_Rollup['arr_flights'])*100
df2_AirlineDelay_diverted = (df2_AirlineDelay_Status_Rollup['arr_diverted'] / df2_AirlineDelay_Status_Rollup['arr_flights'])*100

####

df2_AirlineDelay_Status_Rollup["% Delayed"] = df2_AirlineDelay_ArrDel15 
df2_AirlineDelay_Status_Rollup["% Cancelled"] = df2_AirlineDelay_ArrCancel 
df2_AirlineDelay_Status_Rollup["% Diverted"] = df2_AirlineDelay_diverted

print(f"---------------------Summary of Delayed, Cancelled and Diverted Flights in US (2003 to 2022)-----------")
df2_AirlineDelay_Status_Rollup
# df_AirlineDelayNASConvertToHrs= df_AirlineDelayTotalMins5CarrMaxYr['nas_delay']/60
# df_AirlineDelayTotalMins5CarrMaxYr["arr_delay_hrs"] = df_AirlineDelayArrConvertToH

#Table 1
```
